```python
from skimage.morphology import (erosion, dilation, opening, closing, white_tophat)
from skimage.morphology import black_tophat, skeletonize, convex_hull_image

footprint = [[0, 1, 0], [1, 1, 1],[0, 1, 0]]

print("Structuring element:\n", footprint)
eroded = erosion(img, footprint)

print("Original Img:\n", img)
print("Eroded Img:\n", eroded)
plt.imshow(eroded)
```

```python
# image morphology to be covered
ero = erosion(img, footprint)
# Pixel (output) = 1 {if FIT}
# Pixel (output) = 0 {otherwise}
print("Eroded Image:\n", ero)

dil = dilation(img, footprint)
# Pixel (output) = 1 {if HIT}
# Pixel (output) = 0 {otherwise}
print("Dilated Image:\n", dil)

open = dilation(erosion(img, footprint), footprint)
print("Opening:\n", open)

close = erosion(dilation(img, footprint), footprint)
print("Closing:\n", close)

tophat = img - open
print("Top hat:\n", tophat)

bothat = img - close
print("bothat:\n", bothat)

morphologicalgradient = dil - ero
print("Morphological grad:\n", morphologicalgradient)
```

```python
new_img = np.zeros((chel.shape[0], chel.shape[1]))
# Grayscale filter
for x in range(chel.shape[0]):
  for y in range(chel.shape[1]):
    # Get original pixel colors
    r, g, b = chel[x, y]

    # New pixel colors
    # gray =  (r+g+b)/3
    gray = 0.299*r + 0.587*g + 0.114*b
    new_img[ x, y] = gray

new_img.shape
plt.imshow(new_img, cmap="gray")
```

```python
# brightness
new_img = np.zeros(chel.shape, dtype=np.uint8)
def truncate(x):
    '''makes sure returned value is between 0 and 255'''
    return min(255, max(0, x))

d =  100
for x in range(chel.shape[0]):
  for y in range(chel.shape[1]):
        r, g, b = chel[x, y]

        # d is the brightness increase
        r_ = truncate(r + d)
        g_ = truncate(g + d)
        b_ = truncate(b + d)

        new_pixel = [r_, g_, b_]
        new_img[ x, y] = new_pixel

plt.imshow(new_img)
```
